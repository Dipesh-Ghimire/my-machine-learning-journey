```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_csv("canada_per_capita_income.csv");
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("PCI(US$)")
plt.scatter(df.year,df.PCI,color='red',marker='+')
```

```python
regobj = linear_model.LinearRegression()
```

```python
regobj.fit(df[['year']],df.PCI)
```

```python
regobj.predict([[1990]])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("PCI(US$)")
plt.scatter(df.year,df.PCI,color='red',marker='+')
plt.plot(df.year,regobj.predict(df[['year']]),color='blue')
```
